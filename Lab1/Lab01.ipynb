### Zad 1:

Zaimplementuj funkcję:
$f(x) = \frac{1}{8}x^{4} + 2(x + 0.7)^{3} - \frac{1}{2}x^{2} - 38 x + 1$.



```python
def f(x):
    return 1/8 * x**4 + 2*(x+0.7)**3 - 1/2*x**2 - 38*x + 1
```

### Zad 2:

Oblicz wartość funkcji $f(x)$ w przedziale [-20, 20] z krokiem $\Delta = 1$. Wyniki zapisz na liście.



```python
result = []

for x in range(-20, 21):
    result.append(f(x))
```

### Zad 3:

Wyświetl wykres funkcji $f(x)$ korzystając z wyników z zad. 2.

```python
print(result)
```

```python
import matplotlib.pyplot as plt

indexes = [x for x in range(-20, 21)]

plt.plot(indexes, result, marker='o')
plt.title("Wyniki")
plt.xlabel("Index")
plt.ylabel("Wartość")
plt.show()
```

### Zad 4:

Korzystając z listy otrzymanej w zad. 2, znajdź takie $x$, dla którego funkcja $f(x)$ przyjmuje najmniejszą wartość. Wyświetl $x$ oraz $f(x)$.

```python
minimum = min(result)
minimum_index = result.index(minimum) - 20

minimum, minimum_index
```

### Zad 5:

Wyznacz inny krok ($\Delta' < \Delta$) z zakresu [-20, 20] i sprawdź czy odnajdziesz mniejszą wartość funkcji $f(x)$ niż w poprzednim zadaniu? Jeżeli tak, to podaj $x$ oraz wartość funkcji $f(x)$.

```python
result_2 = []
indexes_2 = []

for x in range(-20, 21):
    result_2.append(f(x))
    indexes_2.append(x)
    if x != 20:
        result_2.append(f(x + 0.5))
        indexes_2.append(x + 0.5)
```

```python
minimum_2 = min(result_2)
minimum_index_2 = result_2.index(minimum_2) * 0.5 - 20

minimum_2, minimum_index_2
```

### Zad 6:

Wyświetl wykres (parametry wykresu dobierz samodzielnie) prezentujący minimum funkcji $f(x)$ znalezione w zad. 5.


```python
import matplotlib.pyplot as plt

plt.plot(indexes_2, result_2)  # marker='o' by dodać kropki
plt.scatter(minimum_index_2, minimum_2, color='red', label='Minimum')
plt.title("Wyniki")
plt.xlabel("Index")
plt.ylabel("Wartość")
plt.show()
```

### Zad 7:

Wygeneruj $k$ losowych wartości $x$ w przedziale [-20, 20], a następnie znajdź minimum funkcji $f(x)$.
Wyświetl wykres funkcji.

```python
import random

random.seed(67)

k = 100

random_list = sorted([random.uniform(-20, 20) for _ in range(k)])
```

```python
result_random = []

for x in random_list:
    result_random.append(f(x))
```

```python
minimum_random = min(result_random)
minimum_index_random = random_list[result_random.index(minimum_random)]

minimum_random, minimum_index_random
```

```python
plt.plot(random_list, result_random)  # marker='o' by dodać kropki
plt.scatter(minimum_index_random, minimum_random, color='red', label='Minimum')
plt.title("Wyniki")
plt.xlabel("Index")
plt.ylabel("Wartość")
plt.show()
```
