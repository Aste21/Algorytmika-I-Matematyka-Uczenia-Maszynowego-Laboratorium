- uruchomienie komórki zawierającej jedynie nazwę zmiennej (typu `Vector`) spowoduje wyświetlenie nazwy typu `Vector`, adresu zmiennej w pamięci (funkcja `id`) oraz jej zawartości, np. `<Vector([1, 2, 3]) at 140185630442880>` (metoda `__repr__`)

````python
import matplotlib.pyplot as plt

class Vector:
    # ZADANIE 1
    def __init__(self, list_coordinates):
        self.coordinates = list_coordinates
        self.dimension = len(list_coordinates)

    def __str__(self):
        return f"Vector: {self.coordinates}"
    
    def get_modulo(self):
        return sum(x**2 for x in self.coordinates) ** 0.5
    
    def __add__(self, other):
        if self.dimension != other.dimension:
            raise ValueError("Vectors must have the same dimension for addition.")
        return Vector([a + b for a, b in zip(self.coordinates, other.coordinates)])
    
    def __sub__(self, other):
        if self.dimension != other.dimension:
            raise ValueError("Vectors must have the same dimension for subtraction.")
        return Vector([a - b for a, b in zip(self.coordinates, other.coordinates)])
    
    def __repr__(self):
        return f"<Vector({self.coordinates}) at {hex(id(self))}>"
    
    # ZADANIE 2
    def __eq__(self, other):
        if self.dimension != other.dimension:
            raise ValueError("Vectors must have the same dimension for comparison.")
        result = []
        for a, b in zip(self.coordinates, other.coordinates):
            if a == b:
                result.append(True)
            else:
                result.append(False)
        return Vector(result)
    
    def __ne__(self, other):
        if self.dimension != other.dimension:
            raise ValueError("Vectors must have the same dimension for comparison.")
        result = []
        for a, b in zip(self.coordinates, other.coordinates):
            if a != b:
                result.append(True)
            else:
                result.append(False)
        return Vector(result)

    def __lt__(self, other):
        if self.dimension != other.dimension:
            raise ValueError("Vectors must have the same dimension for comparison.")
        result = []
        for a, b in zip(self.coordinates, other.coordinates):
            if a < b:
                result.append(True)
            else:
                result.append(False)
        return Vector(result)
    
    def __le__(self, other):
        if self.dimension != other.dimension:
            raise ValueError("Vectors must have the same dimension for comparison.")
        result = []
        for a, b in zip(self.coordinates, other.coordinates):
            if a <= b:
                result.append(True)
            else:
                result.append(False)
        return Vector(result)
    
    def __gt__(self, other):
        if self.dimension != other.dimension:
            raise ValueError("Vectors must have the same dimension for comparison.")
        result = []
        for a, b in zip(self.coordinates, other.coordinates):
            if a > b:
                result.append(True)
            else:
                result.append(False)
        return Vector(result)
    
    def __ge__(self, other):
        if self.dimension != other.dimension:
            raise ValueError("Vectors must have the same dimension for comparison.")
        result = []
        for a, b in zip(self.coordinates, other.coordinates):
            if a >= b:
                result.append(True)
            else:
                result.append(False)
        return Vector(result)
    
    # ZADANIE 3
    def display(self):
        if self.dimension != 2:
            raise ValueError("Display method is only implemented for 2D vectors.")
        
        plt.figure()

        plt.arrow(0, 0, self.coordinates[0], self.coordinates[1], head_width=0.1, head_length=0.1, fc='blue', ec='blue')
        plt.show()
````
